```python
def quick_report(df: pd.DataFrame, top_n: int = 10) -> None:
    print("=== Quick report ===")
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
    
    print("\nColumn dtypes:")
    print(df.dtypes)
    
    print("\nMissing values per column:")
    print(df.isna().sum().sort_values(ascending=False).head(top_n))
    
    print("\nSample rows:")
    print(df.head(5))
    df
```

## To Remove special characters

```python
import string

df = df.applymap(
    lambda x: x.translate(str.maketrans('', '', string.punctuation)) if isinstance(x, str) else x
)
df
```

## remove especial character 

```python
import re

df = df.applymap(
    lambda x: re.sub(r'[^A-Za-z0-9 _]+', '', x) if isinstance(x, str) else x
)
df
```
